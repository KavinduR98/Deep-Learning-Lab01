```python
import numpy as np
import matplotlib.pyplot as plt

random_array_100000x1 = np.random.exponential(scale=1, size=(100000, 1))

plt.hist(random_array_100000x1, density=True, bins=50, histtype="step", color="blue", label="Exponential")
plt.hist(np.random.rand(100000), density=True, bins=50, histtype="step", color="green", label="Uniform")
plt.hist(np.random.randn(100000), density=True, bins=50, histtype="step", color="red", label="Normal")
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z, cmap="viridis")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.title("3D Plot of Z = X**2 + Y**2")
plt.show()
```

```python
import pandas as pd

# Create a DataFrame with the provided data
data = {
    "Name": ["Bulbasaur", "Ivysaur", "Venusaur", "Charmander", "Charmeleon"],
    "Type 1": ["Grass", "Grass", "Grass", "Fire", "Fire"],
    "Type 2": ["Poison", "Poison", "Poison", "NaN", "NaN"],
    "Total": [318, 405, 525, 309, 405],
    "HP": [45, 60, 80, 39, 58],
    "Attack": [49, 62, 82, 52, 64],
    "Defense": [49, 63, 83, 43, 58],
    "Sp. Atk": [65, 80, 100, 60, 80],
    "Sp. Def": [65, 80, 100, 50, 65],
    "Speed": [45, 60, 80, 65, 80],
    "Stage": [1, 2, 3, 1, 2],
    "Legendary": [False, False, False, False, False]
}

df = pd.DataFrame(data)

# Display the DataFrame
df

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select the desired features
selected_features = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]

# Calculate Pearson correlation
pearson_corr = df[selected_features].corr(method="pearson")

# Calculate Spearman correlation
spearman_corr = df[selected_features].corr(method="spearman")

# Create subplots for heatmaps
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot Pearson correlation heatmap
sns.heatmap(pearson_corr, annot=True, cmap="coolwarm", center=0, ax=axes[0])
axes[0].set_title("Pearson Correlation Heatmap")

# Plot Spearman correlation heatmap
sns.heatmap(spearman_corr, annot=True, cmap="coolwarm", center=0, ax=axes[1])
axes[1].set_title("Spearman Correlation Heatmap")

plt.tight_layout()
plt.show()
```
